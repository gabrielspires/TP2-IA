```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def plotar(arquivo, label):
    df = pd.read_csv(arquivo, delim_whitespace=True, header=None, names=['Episdio','Recompensa mdia'])
    plt.plot(df['Episdio'], df['Recompensa mdia'], label=label)

plt.figure(figsize=(15,7))
plt.xlabel('Episdio', fontsize=16)
plt.ylabel('Recompensa mdia', fontsize=16)
plt.grid(True)
plt.title('Q-Learning - Caso do labirinto 5x6', fontsize=20)

plotar('rewards/pacmaze1.txt_0.1_0.1_300_300.txt', '=0.1, =0.1')
plotar('rewards/pacmaze1.txt_0.1_0.5_300_300.txt', '=0.1, =0.5')
plotar('rewards/pacmaze1.txt_0.1_0.9_300_300.txt', '=0.1, =0.9')
plotar('rewards/pacmaze1.txt_0.5_0.1_300_300.txt', '=0.5, =0.1')
plotar('rewards/pacmaze1.txt_0.5_0.5_300_300.txt', '=0.5, =0.5')
plotar('rewards/pacmaze1.txt_0.5_0.9_300_300.txt', '=0.5, =0.9')
plotar('rewards/pacmaze1.txt_0.9_0.1_300_300.txt', '=0.9, =0.1')
plotar('rewards/pacmaze1.txt_0.9_0.5_300_300.txt', '=0.9, =0.5')
plotar('rewards/pacmaze1.txt_0.9_0.9_300_300.txt', '=0.9, =0.9')
plt.legend()
plt.savefig('doc/grafico1.pdf',format='pdf', bbox_inches='tight')
plt.show()


plt.figure(figsize=(15,7))
plt.xlabel('Episdio', fontsize=16)
plt.ylabel('Recompensa mdia', fontsize=16)
plt.grid(True)
plt.title('Q-Learning - Caso do labirinto 7x14', fontsize=20)

plotar('rewards/pacmaze2.txt_0.1_0.1_1000_1000.txt', '=0.1, =0.1')
plotar('rewards/pacmaze2.txt_0.1_0.5_1000_1000.txt', '=0.1, =0.5')
plotar('rewards/pacmaze2.txt_0.1_0.9_1000_1000.txt', '=0.1, =0.9')
plotar('rewards/pacmaze2.txt_0.5_0.1_1000_1000.txt', '=0.5, =0.1')
plotar('rewards/pacmaze2.txt_0.5_0.5_1000_1000.txt', '=0.5, =0.5')
plotar('rewards/pacmaze2.txt_0.5_0.9_1000_1000.txt', '=0.5, =0.9')
plotar('rewards/pacmaze2.txt_0.9_0.1_1000_1000.txt', '=0.9, =0.1')
plotar('rewards/pacmaze2.txt_0.9_0.5_1000_1000.txt', '=0.9, =0.5')
plotar('rewards/pacmaze2.txt_0.9_0.9_1000_1000.txt', '=0.9, =0.9')
plt.legend()
plt.savefig('doc/grafico2.pdf',format='pdf', bbox_inches='tight')
plt.show()


plt.figure(figsize=(15,7))
plt.xlabel('Episdio', fontsize=16)
plt.ylabel('Recompensa mdia', fontsize=16)
plt.grid(True)
plt.title('Q-Learning - Caso do labirinto 14x20', fontsize=20)

plotar('rewards/pacmaze3.txt_0.1_0.1_30000_30000.txt', '=0.1, =0.1')
plotar('rewards/pacmaze3.txt_0.1_0.5_30000_30000.txt', '=0.1, =0.5')
plotar('rewards/pacmaze3.txt_0.1_0.9_30000_30000.txt', '=0.1, =0.9')
plotar('rewards/pacmaze3.txt_0.5_0.1_30000_30000.txt', '=0.5, =0.1')
plotar('rewards/pacmaze3.txt_0.5_0.5_30000_30000.txt', '=0.5, =0.5')
plotar('rewards/pacmaze3.txt_0.5_0.9_30000_30000.txt', '=0.5, =0.9')
plotar('rewards/pacmaze3.txt_0.9_0.1_30000_30000.txt', '=0.9, =0.1')
plotar('rewards/pacmaze3.txt_0.9_0.5_30000_30000.txt', '=0.9, =0.5')
plotar('rewards/pacmaze3.txt_0.9_0.9_30000_30000.txt', '=0.9, =0.9')
plt.legend()
plt.savefig('doc/grafico3.pdf',format='pdf', bbox_inches='tight')
plt.show()


plt.figure(figsize=(15,7))
plt.xlabel('Episdio', fontsize=16)
plt.ylabel('Recompensa mdia', fontsize=16)
plt.grid(True)
plt.title('Q-Learning - Caso do labirinto 11x20', fontsize=20)

plotar('rewards/pacmaze4.txt_0.1_0.1_30000_30000.txt', '=0.1, =0.1')
plotar('rewards/pacmaze4.txt_0.1_0.5_30000_30000.txt', '=0.1, =0.5')
plotar('rewards/pacmaze4.txt_0.1_0.9_30000_30000.txt', '=0.1, =0.9')
plotar('rewards/pacmaze4.txt_0.5_0.1_30000_30000.txt', '=0.5, =0.1')
plotar('rewards/pacmaze4.txt_0.5_0.5_30000_30000.txt', '=0.5, =0.5')
plotar('rewards/pacmaze4.txt_0.5_0.9_30000_30000.txt', '=0.5, =0.9')
plotar('rewards/pacmaze4.txt_0.9_0.1_30000_30000.txt', '=0.9, =0.1')
plotar('rewards/pacmaze4.txt_0.9_0.5_30000_30000.txt', '=0.9, =0.5')
plotar('rewards/pacmaze4.txt_0.9_0.9_30000_30000.txt', '=0.9, =0.9')
plt.legend()
plt.savefig('doc/grafico4.pdf',format='pdf', bbox_inches='tight')
plt.show()
```
