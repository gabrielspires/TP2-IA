```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def plotar(arquivo, label):
    df = pd.read_csv(arquivo, delim_whitespace=True, header=None, names=['Episódio','Recompensa média'])
    plt.plot(df['Episódio'], df['Recompensa média'], label=label)

plt.figure(figsize=(15,7))
plt.xlabel('Episódio')
plt.ylabel('Recompensa média')
plt.grid(True)
plt.title('Q-Learning - Caso do labirinto 5x6')

plotar('rewards/pacmaze1.txt_0.1_0.1_300_300.txt', 'α=0.1, ε=0.1')
plotar('rewards/pacmaze1.txt_0.1_0.5_300_300.txt', 'α=0.1, ε=0.5')
plotar('rewards/pacmaze1.txt_0.1_0.9_300_300.txt', 'α=0.1, ε=0.9')
plotar('rewards/pacmaze1.txt_0.5_0.1_300_300.txt', 'α=0.5, ε=0.1')
plotar('rewards/pacmaze1.txt_0.5_0.5_300_300.txt', 'α=0.5, ε=0.5')
plotar('rewards/pacmaze1.txt_0.5_0.9_300_300.txt', 'α=0.5, ε=0.9')
plotar('rewards/pacmaze1.txt_0.9_0.1_300_300.txt', 'α=0.9, ε=0.1')
plotar('rewards/pacmaze1.txt_0.9_0.5_300_300.txt', 'α=0.9, ε=0.5')
plotar('rewards/pacmaze1.txt_0.9_0.9_300_300.txt', 'α=0.9, ε=0.9')
plt.legend()
plt.savefig('doc/grafico1.pdf',format='pdf')
plt.show()


plt.figure(figsize=(15,7))
plt.xlabel('Episódio')
plt.ylabel('Recompensa média')
plt.grid(True)
plt.title('Q-Learning - Caso do labirinto 7x14')

plotar('rewards/pacmaze2.txt_0.1_0.1_1000_1000.txt', 'α=0.1, ε=0.1')
plotar('rewards/pacmaze2.txt_0.1_0.5_1000_1000.txt', 'α=0.1, ε=0.5')
plotar('rewards/pacmaze2.txt_0.1_0.9_1000_1000.txt', 'α=0.1, ε=0.9')
plotar('rewards/pacmaze2.txt_0.5_0.1_1000_1000.txt', 'α=0.5, ε=0.1')
plotar('rewards/pacmaze2.txt_0.5_0.5_1000_1000.txt', 'α=0.5, ε=0.5')
plotar('rewards/pacmaze2.txt_0.5_0.9_1000_1000.txt', 'α=0.5, ε=0.9')
plotar('rewards/pacmaze2.txt_0.9_0.1_1000_1000.txt', 'α=0.9, ε=0.1')
plotar('rewards/pacmaze2.txt_0.9_0.5_1000_1000.txt', 'α=0.9, ε=0.5')
plotar('rewards/pacmaze2.txt_0.9_0.9_1000_1000.txt', 'α=0.9, ε=0.9')
plt.legend()
plt.savefig('doc/grafico2.pdf',format='pdf')
plt.show()


plt.figure(figsize=(15,7))
plt.xlabel('Episódio')
plt.ylabel('Recompensa média')
plt.grid(True)
plt.title('Q-Learning - Caso do labirinto 14x20')

plotar('rewards/pacmaze3.txt_0.1_0.1_30000_30000.txt', 'α=0.1, ε=0.1')
plotar('rewards/pacmaze3.txt_0.1_0.5_30000_30000.txt', 'α=0.1, ε=0.5')
plotar('rewards/pacmaze3.txt_0.1_0.9_30000_30000.txt', 'α=0.1, ε=0.9')
plotar('rewards/pacmaze3.txt_0.5_0.1_30000_30000.txt', 'α=0.5, ε=0.1')
plotar('rewards/pacmaze3.txt_0.5_0.5_30000_30000.txt', 'α=0.5, ε=0.5')
plotar('rewards/pacmaze3.txt_0.5_0.9_30000_30000.txt', 'α=0.5, ε=0.9')
plotar('rewards/pacmaze3.txt_0.9_0.1_30000_30000.txt', 'α=0.9, ε=0.1')
plotar('rewards/pacmaze3.txt_0.9_0.5_30000_30000.txt', 'α=0.9, ε=0.5')
plotar('rewards/pacmaze3.txt_0.9_0.9_30000_30000.txt', 'α=0.9, ε=0.9')
plt.legend()
plt.savefig('doc/grafico3.pdf',format='pdf')
plt.show()


plt.figure(figsize=(20,7))
plt.xlabel('Episódio')
plt.ylabel('Recompensa média')
plt.grid(True)
plt.title('Q-Learning - Caso do labirinto 11x20')

plotar('rewards/pacmaze4.txt_0.1_0.1_30000_30000.txt', 'α=0.1, ε=0.1')
plotar('rewards/pacmaze4.txt_0.1_0.5_30000_30000.txt', 'α=0.1, ε=0.5')
plotar('rewards/pacmaze4.txt_0.1_0.9_30000_30000.txt', 'α=0.1, ε=0.9')
plotar('rewards/pacmaze4.txt_0.5_0.1_30000_30000.txt', 'α=0.5, ε=0.1')
plotar('rewards/pacmaze4.txt_0.5_0.5_30000_30000.txt', 'α=0.5, ε=0.5')
plotar('rewards/pacmaze4.txt_0.5_0.9_30000_30000.txt', 'α=0.5, ε=0.9')
plotar('rewards/pacmaze4.txt_0.9_0.1_30000_30000.txt', 'α=0.9, ε=0.1')
plotar('rewards/pacmaze4.txt_0.9_0.5_30000_30000.txt', 'α=0.9, ε=0.5')
plotar('rewards/pacmaze4.txt_0.9_0.9_30000_30000.txt', 'α=0.9, ε=0.9')
plt.legend()
plt.savefig('doc/grafico4.pdf',format='pdf')
plt.show()
```
